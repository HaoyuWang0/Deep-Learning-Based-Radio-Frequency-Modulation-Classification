# Results Visualizations

```python
# Models:
#
# RADIOML 2016.10A:
#     CNN_2016
#     LSTM_2016
#     CNN-LSTM_2016
#
# RADIOML 2018.10A:
#     ResNet33_2018
#     LSTM_2018
#     CNN-LSTM_2018
#
# For each model, we plot its Test Accuracy vs. SNRs. 
# To run this file, please make sure all .npz files are under the same directory.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pickle

# Figure size
plt.rcParams['figure.figsize'] = [10, 6]

def plot_acc_vs_snrs(filename,title):
    '''
    Plot the test accuracy vs. SNRs.
    :param filename: name of data file   :type: str
    :param title: title of the plot      :type: str
    :return: none
    '''
    
    # Load results from npz files
    results = np.load(filename)['arr_0']
    
    # Get snrs and accuracies
    SNRs = results[:,0]
    Accs = results[:,1]
    
    # Plot the graph
    plt.plot(SNRs,Accs)
    plt.title(title)
    plt.xlabel('SNR[db]')
    plt.ylabel('Accuracy')
    
    ax = plt.gca()
    ax.set_ylim([0, 100])
    
```

## CNN 2016 Results

```python
plot_acc_vs_snrs('CNN_2016.npz','CNN 2016')
```

## LSTM 2016 Results

```python
plot_acc_vs_snrs('LSTM_2016.npz','LSTM 2016')
```

## CNN-LSTM 2016 Results

```python
plot_acc_vs_snrs('CNNLSTM_2016.npz','CNN-LSTM 2016')
```

## ResNet33 2018 Results

```python
plot_acc_vs_snrs('ResNet33_2018.npz','ResNet33 2018')
```

## LSTM 2018 Results

```python
plot_acc_vs_snrs('LSTM_2018.npz','LSTM 2018')
```

## CNN-LSTM 2018 Results

```python
plot_acc_vs_snrs('CNNLSTM_2018.npz','CNN-LSTM 2018')
```
